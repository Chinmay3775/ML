```python
import pandas as pd
```

```python
df= pd.read_csv('htwt.csv')
df
```

```python
df.shape
```

```python
df.describe()
```

```python
x=df.iloc[:50,[1,2]].values
x
```

```python
x.ndim
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=3)
model.fit(x)
```

```python
y_pred =model.predict(x)
y_pred
```

```python
x[y_pred==0,0]
```

```python
x[y_pred==0,1]
```

```python
x[y_pred==1,0]
```

```python
x[y_pred==1,1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=10,c='red' ,label='cluster 1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=10,c='g',label='cluster 2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=10,c='b',label='cluster 3')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],c='yellow',s=150)
plt.legend()
plt.show()
```
